```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader,random_split
from matplotlib import pyplot as plt
```

```python
# Extract the path of folder.
path = os.getcwd()
path = os.path.join(path,'FRUIT-16K')

# 1. Define the path to the main directory that contains all your image folders
data_dir = path # <-- Change this path to your actual directory
```

```python
# 2. Data loading for validation of a model.
valadation_data = transforms.Compose([
    transforms.Resize(256),                  # Resize the smaller edge to 256
    transforms.CenterCrop(224),              # Crop the center to 224x224
    transforms.ToTensor(),                   # Convert to tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]) # Normalize
])

# Data augimentation for traning of data.
train_transform = transforms.Compose([
    transforms.RandomResizedCrop(224),       # Random crop and resize
    transforms.RandomHorizontalFlip(),       # Randomly flip the image horizontally
    transforms.RandomRotation(15),           # Randomly rotate by +/- 15 degrees
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1), # Jitter colors
    transforms.ToTensor(),                   # Convert to tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]) # Normalize
])
```

```python
full_dataset = datasets.ImageFolder(root=data_dir, transform=valadation_data)
```

```python
# --- 3. Split the dataset programmatically ---
total_size = len(full_dataset)
train_size = int(0.8 * total_size) # 80% for training
val_size = total_size - train_size  # Remaining 20% for validation
```

```python
train_dataset_subset, val_dataset_subset = random_split(
    full_dataset, 
    [train_size, val_size],
    generator=torch.Generator().manual_seed(42) # Use a seed for reproducibility
)
```

```python
train_loader = DataLoader(
    train_dataset_subset,
    batch_size=32,
    shuffle=True, # Shuffle training data
    num_workers=4
)

val_loader = DataLoader(
    val_dataset_subset,
    batch_size=32,
    shuffle=False, # Do not shuffle validation data (for consistent evaluation)
    num_workers=4
)

```

```python
print(f"Total images found: {total_size}")
print(f"Training images: {len(train_dataset_subset)}")
print(f"Validation images: {len(val_dataset_subset)}")
print(f"Classes: {full_dataset.classes}")
```

```python
import torchvision
import numpy as np
def imshow(inp, title=None):
    """Imshow for Tensor."""
    # Move tensor from GPU to CPU if necessary, and convert to numpy array
    inp = inp.cpu().numpy().transpose((1, 2, 0))
    
    # Denormalize the image (reverse the normalization applied in transforms)
    # These are the standard ImageNet mean and std values we used
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    
    # Clip values to the valid [0, 1] range expected by matplotlib
    inp = np.clip(inp, 0, 1)
    
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated

# Get a batch of training data
# Note: 'train_loader' might be named 'dataloader' in your initial code block
inputs, classes = next(iter(train_loader))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs[:4]) # Show the first 4 images in the batch

# Get the corresponding class names
class_names = full_dataset.classes # Using the class names from the full dataset object

imshow(out, title=[class_names[x] for x in classes[:4]])

# Function to visualize multiple images in a grid format
def visualize_data_batch(dataloader, classes_list, num_images=8):
    # Get a batch of inputs and labels
    inputs, labels = next(iter(dataloader))
    
    # Create a grid of images
    out = torchvision.utils.make_grid(inputs[:num_images])

    # Display the grid
    imshow(out, title=[classes_list[x] for x in labels[:num_images]])
    plt.show() # Keep the plot open until closed manually

# Call the visualization function
visualize_data_batch(train_loader, full_dataset.classes, num_images=8)
```
